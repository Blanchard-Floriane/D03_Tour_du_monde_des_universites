### Questions du manager :

```python
#give me every unique value of the column "year" for each data frame
years_saudi_arabia_ranking = saudi_arabia_ranking["year"].unique()
years_shangai_ranking = shangai_ranking["year"].unique()
years_times_ranking = times_ranking["year"].unique()

print("Classement d'Arabie Saoudite :", years_saudi_arabia_ranking)
print("Classement Shangaï :", years_shangai_ranking)
print("Classement du Times :", years_times_ranking)
```

```python
arabia_univ_av = saudi_arabia_ranking["year"].value_counts().mean().round(2)
shangai_univ_av = shangai_ranking["year"].value_counts().mean().round(2)
times_univ_av = times_ranking["year"].value_counts().mean().round(2)

print("Le classement du Center for WUR prend en compte", arabia_univ_av, "universités en moyenne annuellement.")
print("Le classement du Shangai Ranking prend en compte", shangai_univ_av, "universités en moyenne annuellement.")
print("Le classement du Times prend en compte", times_univ_av, "universités en moyenne annuellement.")
```

```python
arabia_strasbourg_ranking = saudi_arabia_ranking.loc[saudi_arabia_ranking["institution"].str.contains("Strasbourg"), ["year", "institution", "world_rank"]]

#here "university_name" containt some Na/NaN, so this stop the function .loc. Str.contains("....", na=False) will tell to the program to understand Na but not count this row. 
shangai_strasbourg_ranking = shangai_ranking.loc[shangai_ranking["university_name"].str.contains("Strasbourg", na=False), ["year", "university_name", "world_rank"]]

times_strasbourg_ranking = times_ranking.loc[times_ranking["university_name"].str.contains("Strasbourg"), ["year", "university_name", "world_rank"]]

display("Center for WUR :", arabia_strasbourg_ranking)
display("Shangai Ranking :", shangai_strasbourg_ranking)
display("Times Higher Education World University Ranking :", times_strasbourg_ranking)
```
